Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Эффективность для признака "rougher.output.recovery"
- Проверим правильность вычисления эффективности обогащения чернового концентрата на обучающей выборке для признака "rougher.output.recovery".
- Сначала удалим пропуски из столбцов.
- Найдем MAE между нашими расчетами и значениями признака. Опишем выводы.




Расчитаем эффективность с помощью формулы recovery (добвлена в начало проекта). Заполним самые низкие и высокие значения numpy nan

concentrate — золото в концентрате после флотации: rougher.output.concentrate_au <br>
feed — золото в концентрате до флотации: rougher.input.feed_au <br>
tail — золото в отвальных хвостах после флотации: rougher.output.tail_au

```python
def recovery(concentrate, feed, tail):
    prerecovery = (concentrate * (feed - tail)) / (feed * (concentrate - tail))
    recovery = prerecovery * 100

    recovery[recovery<0] = np.nan
    recovery[recovery>100] = np.nan
    
    return recovery

concentrate = gr_train['rougher.output.concentrate_au']
feed = gr_train['rougher.input.feed_au']
tail = gr_train['rougher.output.tail_au']

recovery_df = recovery(concentrate, feed, tail)
```

Расчитаем rougher_recovery на обучающем датасете:

```python
recovery = recovery(gr_train['rougher.output.concentrate_au'], gr_train['rougher.input.feed_au'], gr_train['rougher.output.tail_au'])
print(recovery.describe())
```

Заменим все пропущенные значения на 0 и найдем среднее абсолютное отклонение, MAE

```python
mae_recovery = mean_absolute_error(gr_train['rougher.output.recovery'].fillna(0),recovery_df.fillna(0))
print('Среднее абсолютное отклонение равно', mae_recovery)
```

#### ВЫВОД:
Среднее абсолютное отклонение, MAE, иммеет малое отклонение. Следовательно, значение эффективности в данных првильное.

## Модель

### Введение новых функций
Напишем две функции по формулам, которые даны в описании проекта: 
- sMAPE
- Итоговой sMAPE 

```python
def sMAPE(target, predictions):
    up = np.abs(target - predictions)
    down = (np.abs(target) + np.abs(predictions)) / 2 + 1e-15
    return 1 / len(target) * np.nansum(up / down) * 100 
    
    
def final_sMAPE(target, predictions):
    target_rougher = target['rougher.output.recovery']
    target_final = target['final.output.recovery']
    
    predictions_rougher = predictions[:, 0]
    predictions_final = predictions[:, 1]
    
    return 0.25 * sMAPE(target_rougher, predictions_rougher) + 0.75 * sMAPE(target_final, predictions_final)

sMAPE_score = make_scorer(final_sMAPE, greater_is_better=False)
```

### Обучение моделей
Обучим разные модели и оценим их качество кросс-валидацией. <br>
